```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'],
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'],
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
df.replace('Missing', np.nan, inplace=True)
df.replace('NA', np.nan, inplace=True)
```

**df.dropna()** removes rows containing any NaN (missing) values from your DataFrame and returns a new DataFrame with those rows dropped — it doesn't modify df in place unless you pass inplace=True.

by default it has 2 arguments : axis = 'index' , how = 'any'

axis = 'index' ==> drop rows

axis = 'column' ==> drop columns

how = 'any' ==> 1 or more

how = 'all' ==> all row or column values are Nan
