# Video 6 : Add/Remove Rows and Columns From DataFrames

```python
data = {
    "first": ["John", "Alice", "Bob" , "Amina"],
    "last": ["Smith", "Johnson", "Brown" , "Brown"],
    "email": [
        "john.smith@example.com",
        "alice.johnson@example.com",
        "bob.brown@example.com",
        "amina.brown@example.com"
    ]
}
df6 = pd.DataFrame(data)
df6
```

```python
df6['full_name'] = df6['first'] + ' ' + df6['last']
```

```python
df6
```

```python
df6.drop(columns=['first' , 'last'] , inplace=True)
```

```python
df6
```

```python
df6['full_name'].str.split(' ')
# give us list : [John,Smith]
```

```python
df6['full_name'].str.split(' ', expand=True)
# give us list : 2 columns separated instead of list
```

```python
df6
```

```python
df6[['first','last']] = df6['full_name'].str.split(' ', expand=True)
```

```python
df6
```

```python
pd.__version__
```

```python
df6
```

```python
#add row
new_row = pd.DataFrame([{'first': 'Youssef', 'last': 'Kennani'}])
df6 = pd.concat([df6, new_row], ignore_index=True)
```
