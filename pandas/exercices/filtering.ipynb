Filtering - Using Conditionals to Filter Rows and Columns


**Beginner Exercises**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Ami'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rab at', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'years_experience': [1, 8, 4, 15, 0.5, 10, 3]
}
df = pd.DataFrame(data)
```

```python
df
```

Filter people older than 30.

```python
df
```

```python
older_30 = (df['age'] > 30)
df[older_30]
```

Filter people who live in 'Rabat'.

```python
rabat_filt = (df['city'] == 'Rabat')
df[rabat_filt]
```

Filter people with salary >= 6000 AND years_experience > 2.

```python
salary_filt = (df['salary'] >= 6000) & (df['years_experience'] > 2)
df[salary_filt]
```

Filter people who live in either 'Fes' or 'Marrakech' (use isin).

```python
cities = ['Fes' , 'Marrakech']
cities_filt = df['city'].isin(cities)
df[cities_filt]
```

Filter people who do NOT live in 'Casablanca'.



```python
#Methode 1
#df[df['city'] != 'Casablanca']
#Methode 2
is_casa = df['city'] == 'Casablanca'
df[~is_casa]
```

**Intermediate Exercises**

Get the name and salary of everyone earning more than 7000. Don't return the whole dataframe — only those two columns.

```python
df.head(3)
```

```python
salary_filt2 = (df['salary'] > 7000)
df.loc[salary_filt2, ['name' , 'salary']]
```

Filter people whose name contains the letter 'a' (case-insensitive). This introduces .str methods, which you'll use constantly for messy text data (emails, product names, free-text client fields, etc.).

```python
# This searches for 'a' OR 'A' — ignores case on BOTH sides
name_filt = df['name'].str.contains('a', case=False)
df[name_filt]
```

A client emails you: "Send me everyone based in Rabat or Fes, who has more than 3 years of experience, but I only need their name and city — not their salary, that's confidential."


```python
cities = ['Rabat','Fes']
client_filt = ( df['city'].isin(cities) ) & ( df['years_experience'] > 3 )
df[client_filt]

df.loc[client_filt, ['name' , 'city'] ]
```

Challenge Problem — "The Messy Client Spreadsheet"

```python
import pandas as pd
import numpy as np

client_data = {
    'name': ['Yasmine ', 'omar', 'SARA', 'Khalid', 'imane', 'Hamza', 'Lina', 'Youssef'],
    'age': [23, 35, 29, 41, 22, 38, 27, np.nan],
    'city': ['Rabat', 'casablanca', 'Rabat ', 'Fes', 'Marrakech', 'CASABLANCA', np.nan, 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300, 6800],
    'years_experience': [1, 8, 4, 15, 0.5, 10, 3, 6],
    'status': ['active', 'active', 'inactive', 'active', 'active', 'inactive', 'active', 'active']
}
client_df = pd.DataFrame(client_data)
```

Client's request:


"I need a list of all active employees based in Casablanca or Rabat, who have more than 2 years of experience. Give me their name and salary only. Also — some of my data has messy spacing/capitalization, please make sure that doesn't break your filter."

Your task:


1. Clean the city column so 'casablanca', 'CASABLANCA', 'Rabat ' etc. are all treated consistently (hint: .str.strip() and .str.lower() or .str.title(), chained).

2. Build the filter satisfying all the client's conditions (status, city, experience).

3. Return only name and salary for the matching rows.

***Bonus***: handle the NaN in city gracefully — don't let it crash your filter or accidentally get included/excluded wrongly.

```python
client_df
```

```python
cities = ['Casablanca', 'Rabat']
city_filt = client_df['city'].str.strip().str.lower().str.title().isin(cities)
exp_filt = client_df['years_experience'] > 2
active_filt = client_df['status'] == 'active'
goal_filt = city_filt & exp_filt & active_filt
client_df.loc[goal_filt , ['name' , 'salary']]
```
