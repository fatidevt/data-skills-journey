# Pandas Tutorial Exercices For Part 5 :
Updating Rows and Columns - Modifying Data Within DataFrames

**5 Beginner exercises**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'years_experience': [1, 8, 4, 15, 0.5, 10, 3],
    'status': ['active', 'active', 'inactive', 'active', 'active', 'inactive', 'active']
}
df = pd.DataFrame(data)
```

Add a new column called 'salary_mad' that converts salary to MAD (assume current values are in USD, use rate: 1 USD = 10 MAD).

```python
df
```

```python
df['salary_mad'] = df['salary'] * 10
```

```python
df
```

Rename the column 'years_experience' to 'experience' (cleaner for a client deliverable).

```python
df = df.rename( columns = { 'years_experience' : 'experience' })
```

```python
df
```

Update a whole column — give everyone a 15% salary raise (update the 'salary' column directly).


```python
df['salary'] = df['salary'] + df['salary'] * 0.15
```

```python
df
```

Update specific rows — Sara just moved to Casablanca. Update only her city to 'Casablanca' using .loc.


```python
df.loc[2, ['city']] = 'Casablanca'
```

```python
df
```

Add a new column 'seniority' that contains 'senior' if experience >= 5, else 'junior'. (Hint: look up np.where() or use a simple approach you know.)



```python
df['seniority'] = df['experience'].apply(lambda x: 'senior' if x >= 5 else 'junior')
```

```python
df
```

The company decides anyone with 'inactive' status who has more than 5 years experience should be reactivated. Update only those rows' status to 'active' using .loc with a condition.


```python
filt = ( df['status'] == 'inactive') & (df['experience'] > 5 )
df[filt]
```

```python
df.loc[filt , 'status'] = 'active'
```

```python
df
```

Sara got a promotion: her salary increases by 20% AND her seniority becomes 'senior'. Update both columns for Sara only in a single .loc line.


```python
df.loc[2, ['salary', 'seniority']] = [ df.loc[2, 'salary'] * 1.20 , 'senior' ]
```

# Challenge Problem — "The HR Data Cleanup"
A healthcare clinic in Rabat sends you their employee spreadsheet with several problems that need fixing before they can run payroll.


```python
import pandas as pd
import numpy as np

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'department': ['nursing', 'ADMIN', 'Pharmacy', 'nursing', 'admin', 'PHARMACY', 'Nursing'],
    'salary': [6000, 4500, 7200, 6000, 4500, 7200, 6000],
    'experience': [8, 2, 5, 1, 3, 9, 4],
    'status': ['active', 'inactive', 'active', 'active', 'inactive', 'active', 'active']
}
df1 = pd.DataFrame(data)
```

```python
df1
```

Fix the department column — it's inconsistently capitalized ('nursing', 'ADMIN', 'Pharmacy' etc.). Standardize all values to title case.


```python
# Methode 1 , better
# df1['department'] = df1['department'].str.title()
df1['department'] = [x.title() for x in df1['department']]
```

```python
df1
```

Give all Pharmacy staff a 10% salary raise (they received a new certification).


```python
filt = ( df1['department'] == 'Pharmacy' )
df1[filt]
```

```python
df1.loc[filt, 'salary'] = df1.loc[filt, 'salary'] * 1.1
```

```python
df1
```

Reactivate all inactive employees in Admin — the clinic reviewed their cases and cleared them.


```python
filt1 = ( df1['department'] == 'Admin') & ( df1['status'] == 'inactive' )
df1[filt1]
```

```python
df1.loc[filt1 , 'status'] = 'active'
```

```python
df1
```

Add a 'grade' column using these rules:

'A' if experience >= 7
'B' if experience >= 4
'C' if experience < 4

```python
df1['grade'] = df1['experience'].apply(lambda x: 'A' if x >= 7 else ( 'B' if x >= 4 else 'C') )
```

```python
df1
```

Rename 'status' to 'employment_status' for the final client deliverable.



```python
df1 = df1.rename(columns = { 'status' : 'employment_status' })
```

```python
df1
```
