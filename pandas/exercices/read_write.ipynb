```python
import pandas as pd
import sqlite3

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'department': ['Nursing', 'Admin', 'Pharmacy', 'Nursing', 'Admin', 'Pharmacy', 'Nursing'],
    'salary': [6000, 4500, 7200, 6000, 4500, 7200, 6000],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Casablanca', 'Rabat', 'Fes'],
    'experience': [8, 2, 5, 1, 6, 9, 3]
}
df2 = pd.DataFrame(data)
```

1. Save df as employees.json, read it back into df_json, verify shape matches.


```python
df2.to_json('employees.json')
df_json = pd.read_json('employees.json')
df_json
```

2. Write df to SQLite database clinic.db as table 'employees', then read it back with pd.read_sql().


```python
import sqlite3

# Connect (creates file if doesn't exist)
conn = sqlite3.connect('clinic.db')

# Write DataFrame to SQL table
df.to_sql('employees', conn, if_exists='replace', index=False)

# Read back with SQL query — same as MySQL
df_sql = pd.read_sql('SELECT * FROM employees', conn)

df_sql

# Always close connection
conn.close()
df_sql
```

Ex 3 — Simple Explanation
The scenario: a client sends you one Excel file with two sheets inside it:

Sheet 1: senior_staff (experience >= 5)

Sheet 2: junior_staff (experience < 5)

You need to:

Read both sheets
Combine them into one DataFrame
Add a column showing which level each person is
Save as one clean CSV

```python
# Step 1 — read each sheet separately
df_senior = pd.read_excel('staff.xlsx', sheet_name='senior_staff')
df_junior = pd.read_excel('staff.xlsx', sheet_name='junior_staff')

# Step 2 — add level column BEFORE combining
df_senior['level'] = 'senior'
df_junior['level'] = 'junior'

# Step 3 — combine into one DataFrame
df_full = pd.concat([df_senior, df_junior], ignore_index=True)

# Step 4 — save as CSV
df_full.to_csv('full_staff.csv', index=False)
```
