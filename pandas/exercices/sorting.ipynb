**Beginner Exercices**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'experience': [1, 8, 4, 15, 0.5, 10, 3]
}
df = pd.DataFrame(data)
```

Sort the DataFrame by salary from lowest to highest.

```python
df.sort_values(by='salary' , ascending = True, inplace=True)
df
```

Sort by age from highest to lowest.


```python
df.sort_values(by='age', ascending = False , inplace=True)
df
```

Sort by city alphabetically (A→Z).


```python
# no need to write ascending = True cause its by default
df.sort_values(by='city', ascending = True , inplace=True)
df
```

Sort by city A→Z, then by salary highest to lowest within each city.


```python
df.sort_values(by=['city','salary'] , ascending=[True, False], inplace = True)
df
```

Sort by experience descending and reset the index cleanly.



```python
df.sort_values(by=['experience'] , ascending = False , inplace = True )
df
df = df.reset_index(drop=True)
df
```

**Intermediate Exercises**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'experience': [1, 8, 4, 15, 0.5, 10, 3]
}
df3 = pd.DataFrame(data)
```

```python
df3
```

Sort + filter combined

Get the top 3 highest paid employees only — no hardcoding index numbers, use sorting + slicing.

```python
df3.sort_values(by='salary', ascending = False, inplace = True)
df3.head(3)
```

Sort within a filtered group

Filter only employees from Rabat or Casablanca, then sort them by experience descending. Reset the index cleanly.

```python
cities = ['Casablanca', 'Rabat']
filt = df3['city'].isin(cities)
df3 = df3[filt]
df3.sort_values(by='experience', ascending = False, inplace=True)
df3 = df3.reset_index(drop=True)
df3
```

The "client ranking" scenario

A logistics company wants a ranked leaderboard of their drivers:

Sort by experience descending, then salary descending
Add a 'rank' column (1 = most experienced/highest paid)
Only show name, rank, experience, salary

```python
df3.sort_values(['experience', 'salary'] , ascending= [False, False] , inplace= True)
df3 = df3.reset_index(drop = True)
df3['rank'] = df3.index + 1
df3 = df3[['name' , 'rank' , 'experience', 'salary']]
df3
```

# Challenge Problem — "The E-commerce Sales Report"

A Moroccan e-commerce platform wants a clean ranked sales report before their monthly board meeting.


```python
import pandas as pd

data = {
    'product': ['Laptop', 'Hijab', 'Dates', 'Phone', 'Olive Oil', 'Notebook', 'Headphones', 'Argan Oil'],
    'category': ['Electronics', 'Fashion', 'Food', 'Electronics', 'Food', 'Stationery', 'Electronics', 'Beauty'],
    'units_sold': [45, 320, 180, 92, 210, 75, 60, 140],
    'price_per_unit': [8500, 120, 85, 4200, 95, 35, 650, 380],
    'stock_remaining': [10, 45, 200, 5, 80, 150, 8, 60],
    'rating': [4.5, 4.8, 4.2, 4.6, 4.3, 3.9, 4.1, 4.7]
}
df4 = pd.DataFrame(data)
df4
```

Client instructions:

"1. Remove the Stationery category — we're discontinuing it.

2. Sort by category A→Z, then by units_sold highest to lowest within each category.

3. Add a total_revenue column (units_sold × price_per_unit).

4. Add a 'stock_status' column — 'low' if stock_remaining < 20, else 'ok'.

5. Sort the final report by total_revenue descending.

6. Add a rank column (1 = highest revenue).

7. Reset index and show only: rank, product, category, total_revenue, stock_status."

```python
filt = df4['category'] == 'Stationery'
df4[filt]
df4.drop(index = df4[filt].index, inplace = True)
df4
```

```python
df4.sort_values(by=['category', 'units_sold'],  ascending = [True, False],inplace = True)
df4
```

```python
df4['total_revenue'] = df4['units_sold'] * df4['price_per_unit']
df4
```

```python
df4['stock_status'] = df4['stock_remaining'].apply(lambda x: 'low' if x < 20 else 'ok')
df4
```

```python
df4.sort_values('total_revenue' , ascending = False , inplace = True)
df4
```

```python
df4 = df4.reset_index(drop=True)
df4['rank'] = df4.index + 1
df4
```

```python
df4 = df4[['rank','product','category','total_revenue', 'stock_status']]
df4
```
