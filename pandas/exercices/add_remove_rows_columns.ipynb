# Pandas Tutorial Exercices For Part 6 :
Add/Remove Rows and Columns From DataFrames

Beginner Exercices

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'experience': [1, 8, 4, 15, 0.5, 10, 3]
}
df = pd.DataFrame(data)
```

Drop the 'age' column and make the change stick.

```python
df
```

```python
df = df.drop(columns=['age'])
```

Drop the row at index 3 (Khalid) and make the change stick.


```python
df.drop(index=3, inplace = True)
```

```python
df
```

Add a new employee — {'name': 'Youssef', 'city': 'Rabat', 'salary': 6800, 'experience': 6} using pd.concat().


```python
new_row = pd.DataFrame([ {'name': 'Youssef', 'city': 'Rabat', 'salary': 6800, 'experience': 6} ])
df = pd.concat([df, new_row], ignore_index = True )
```

```python
df
```

Drop multiple columns at once — remove both 'city' and 'experience' in a single .drop() call.


```python
df.drop(columns=['city' , 'experience'], inplace = True)
```

```python
df
```

**Intermediate Exercises (3)**

```python
data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina'],
    'age': [23, 35, 29, 41, 22, 38, 27],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Casablanca', 'Rabat'],
    'salary': [4500, 8200, 6100, 9500, 3800, 7200, 5300],
    'experience': [1, 8, 4, 15, 0.5, 10, 3]
}
df = pd.DataFrame(data)
```

```python
df
```

Drop rows based on a condition

Remove all employees with experience < 2 from the DataFrame and reset the index cleanly. Don't hardcode index numbers — use a filter.


```python
filt = df['experience'] < 2
df = df.drop(index=df[filt].index)
df = df.reset_index(drop=True)
```

```python
df
```

Add multiple new rows at once

A new branch just opened. Add these two employees in a single pd.concat() call:

The "client cleanup" scenario

A school admin sends you their teacher dataset and says:

"Remove the 'internal_code' and 'notes' columns — those are confidential. Also remove any teacher with less than 1 year experience, they're still in probation and shouldn't be in this report. Give me a clean, reset index at the end."

```python
teachers = pd.DataFrame({
    'name': ['Ahmed', 'Fatima', 'Karim', 'Souad', 'Bilal'],
    'subject': ['Math', 'Science', 'Arabic', 'English', 'History'],
    'experience': [5, 0.5, 3, 0.3, 8],
    'salary': [8000, 6000, 7500, 6000, 9000],
    'internal_code': ['T001', 'T002', 'T003', 'T004', 'T005'],
    'notes': ['ok', 'probation', 'ok', 'probation', 'ok']
})
df1 = pd.DataFrame(teachers)
```

```python
df1
```

```python
df1  = df1.drop(columns=['internal_code' , 'notes'])
```

```python
df1
```

```python
filt = df1['experience'] < 1
df1 = df1.drop(index=df1[filt].index)
```

```python
df1
```

```python
df1 = df1.reset_index(drop=True)
df1
```

# Challenge Problem — "The Agriculture Export Report"

A Moroccan agricultural export company needs you to clean their product shipment dataset before sending it to an EU client.


```python
import pandas as pd

shipments = pd.DataFrame({
    'product': ['Tomatoes', 'Oranges', 'Olives', 'Dates', 'Peppers', 'Argan Oil'],
    'origin': ['Agadir', 'Marrakech', 'Meknes', 'Erfoud', 'Agadir', 'Essaouira'],
    'weight_kg': [1500, 3200, 800, 950, 2100, 400],
    'price_per_kg': [2.5, 1.8, 6.0, 12.0, 3.2, 45.0],
    'internal_ref': ['A001', 'A002', 'A003', 'A004', 'A005', 'A006'],
    'quality_score': [88, 72, 95, 91, 65, 98],
    'approved': [True, True, False, True, False, True]
})
df2 = pd.DataFrame(shipments)
```

Client instructions:

"1. Remove the internal_ref column — EU clients shouldn't see that.

2. Remove any shipment that is NOT approved.

3. Add two new approved shipments:

{'product': 'Saffron', 'origin': 'Taliouine', 'weight_kg': 50, 'price_per_kg': 850.0, 'quality_score': 99, 'approved': True}

{'product': 'Figs', 'origin': 'Beni Mellal', 'weight_kg': 1100, 'price_per_kg': 4.5, 'quality_score': 83, 'approved': True}

4. Add a 'total_value' column (weight_kg × price_per_kg).

5. Only keep shipments with quality_score >= 80.

6. Reset the index cleanly for the final deliverable."

```python
df2
```

```python
df2.drop(columns=['internal_ref'], inplace=True)
df2
```

```python
filt = df2['approved'] == False
df2 = df2.drop(index=df2[filt].index)
```

```python
df2
```

```python
new_rows = pd.DataFrame([
    {'product': 'Saffron', 'origin': 'Taliouine', 'weight_kg': 50, 'price_per_kg': 850.0, 'quality_score': 99, 'approved': True},
    {'product': 'Figs', 'origin': 'Beni Mellal', 'weight_kg': 1100, 'price_per_kg': 4.5, 'quality_score': 83, 'approved': True}
])
df2 = pd.concat([df2, new_rows], ignore_index=True)
```

```python
df2
```

```python
df2['total_value']=df2['weight_kg'] * df2['price_per_kg']
```

```python
df2
```

```python
filt01 = df2['quality_score'] >= 80
df2 = df2[filt01]
```

```python
df2
```

```python
df2 = df2.reset_index(drop=True)
df2
```
