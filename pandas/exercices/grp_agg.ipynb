# Pandas Tutorial Exercices For Part 8 :
Grouping and Aggregating

# Resume

**one metric**

`df.groupby('city')['salary'].mean() `
             
**multiple at once**

`df.groupby('city')['salary'].agg(['mean','sum','max']) `

**Reset index after groupby**

`df.groupby('city')['salary'].mean().reset_index()`

**Beginner Exercices**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina', 'Youssef'],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Casablanca', 'Fes', 'Rabat', 'Casablanca'],
    'department': ['Nursing', 'Admin', 'Pharmacy', 'Nursing', 'Pharmacy', 'Admin', 'Nursing', 'Pharmacy'],
    'salary': [6000, 4500, 7200, 6000, 7200, 4500, 6000, 7200],
    'experience': [8, 2, 5, 1, 6, 3, 4, 9]
}
df = pd.DataFrame(data)
df
```

Find the average salary per city.

```python
df.groupby('city')['salary'].mean()
```

Find the total salary paid per department.

```python
df.groupby('department')['salary'].sum()
```

Find the maximum experience per city.

```python
df.groupby('city')['experience'].max()
```

Count how many employees per department (hint: .count() or .size()).

```python
#Small note worth knowing:
# .count() counts non-null values
# .size() counts all rows including nulls
df.groupby('department')['name'].count()
```

Find the average and max for salary and experience per city.



```python
# Same metric for all columns
df.groupby('city')[['salary', 'experience']].agg(['mean', 'max'])
```

```python
# Different metrics per column
df.groupby('city').agg(
    salary_mean=('salary', 'mean'),
    salary_max=('salary', 'max'),
    experience_mean=('experience', 'mean'),
    experience_max=('experience', 'max')
)
```

**Intermediate Exercises**

```python
import pandas as pd

data = {
    'name': ['Yasmine', 'Omar', 'Sara', 'Khalid', 'Imane', 'Hamza', 'Lina', 'Youssef'],
    'city': ['Rabat', 'Casablanca', 'Rabat', 'Fes', 'Casablanca', 'Fes', 'Rabat', 'Casablanca'],
    'department': ['Nursing', 'Admin', 'Pharmacy', 'Nursing', 'Pharmacy', 'Admin', 'Nursing', 'Pharmacy'],
    'salary': [6000, 4500, 7200, 6000, 7200, 4500, 6000, 7200],
    'experience': [8, 2, 5, 1, 6, 3, 4, 9]
}
df2 = pd.DataFrame(data)
df2
```

Group by two columns

Find the average salary grouped by both city AND department. Reset the index cleanly.

```python
df2.groupby(['city','department'])['salary'].mean().reset_index()
```

2. Filter after groupby

Find cities where the average salary is above 6000. Return a clean DataFrame with city and average salary only.

(Hint: groupby → reset_index → then filter like normal)

```python
df3 = df2.groupby('city')['salary'].mean().reset_index()
df3
filt = df3['salary'] > 6000
df3[filt]
```

3. The "client summary report" scenario

A healthcare clinic wants a department summary report with:

Total salary paid per department (total_salary)

Average experience per department (avg_experience)

Number of employees per department (headcount)

```python
df2.groupby('department').agg(
    total_salary=('salary', 'sum'),
    avg_experience=('experience', 'mean'),
    headcount=('name', 'count')
).reset_index()
```

# Challenge Problem — "The E-commerce Sales Analysis"

A Moroccan e-commerce platform wants a clean ranked sales report before their monthly board meeting.


```python
import pandas as pd

data = {
    'product': ['Laptop', 'Hijab', 'Dates', 'Phone', 'Olive Oil',
                'Headphones', 'Argan Oil', 'Notebook', 'Charger', 'Honey'],
    'category': ['Electronics', 'Fashion', 'Food', 'Electronics', 'Food',
                 'Electronics', 'Beauty', 'Stationery', 'Electronics', 'Food'],
    'city': ['Rabat', 'Casablanca', 'Fes', 'Rabat', 'Marrakech',
             'Casablanca', 'Rabat', 'Fes', 'Marrakech', 'Rabat'],
    'units_sold': [45, 320, 180, 92, 210, 60, 140, 75, 110, 95],
    'price_per_unit': [8500, 120, 85, 4200, 95, 650, 380, 35, 250, 180],
    'rating': [4.5, 4.8, 4.2, 4.6, 4.3, 4.1, 4.7, 3.9, 4.0, 4.4]
}
df4 = pd.DataFrame(data)
df4
```

**Client instructions:**

"Give me a category performance report with:

Total revenue per category (total_revenue = units_sold × price_per_unit)

Average rating per category (avg_rating)

Total units sold per category (total_units)

Number of products per category (product_count)

```python
df4['revenue'] = df4['units_sold'] * df4['price_per_unit']
df4
rapport1 = df4.groupby('category').agg(
    total_revenue = ('revenue' , 'sum'),
    avg_rating = ('rating' , 'mean'),
    total_units = ('units_sold' , 'sum'),
    product_count = ('product' , 'size')
    ).reset_index()
rapport1
```

**Then:**

Sort by total_revenue descending

Only show categories with total_revenue above 100,000

Reset index and add a rank column

Final columns only: rank, category, total_revenue, avg_rating"



Think before coding — map out your steps first. Some steps must come before others. 💪

```python
rapport1 = rapport1.sort_values(by='total_revenue', ascending = False)
rapport1
```

```python
# filter above 100k
rapport1 = rapport1[rapport1['total_revenue'] > 100000]
rapport1
```

```python
#reset index + add rank
rapport1 = rapport1.reset_index(drop=True)
rapport1['rank'] = rapport1.index + 1
rapport1
```

```python
#final columns
rapport1 = rapport1[['rank', 'category', 'total_revenue', 'avg_rating']]
rapport1
```
