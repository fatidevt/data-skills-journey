**Beginner Exercices**

```python
import pandas as pd
import numpy as np

data = {
    'date': ['2024-01-15', '2024-02-20', '2024-03-10', '2024-04-05',
             '2024-05-18', '2024-06-22', '2024-07-30', '2024-08-14'],
    'product': ['Dates', 'Olive Oil', 'Honey', 'Argan Oil',
                'Dates', 'Honey', 'Olive Oil', 'Argan Oil'],
    'city': ['Rabat', 'Fes', 'Casablanca', 'Rabat',
             'Marrakech', 'Fes', 'Casablanca', 'Rabat'],
    'units_sold': [120, 85, 200, 60, 150, 175, 90, 110],
    'revenue': [1800, 2550, 3000, 4200, 2250, 2625, 2700, 7700]
}
df = pd.DataFrame(data)
df
```

Convert the date column to datetime format.

```python
df['date'] = pd.to_datetime(df['date'])
df
```

Extract year, month, and day into three separate new columns: year, month, day.


```python
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df
```

Extract the day name (Monday, Tuesday...) into a new column called day_name.


```python
df['day_name'] = df['date'].dt.day_name()
df
```

Filter all sales that happened in Q1 (January, February, March).

```python
filt = (df['date'].dt.quarter == 1)
df = df[filt]
df
```

Sort the DataFrame by date ascending and reset the index cleanly.



```python
df = df.sort_values(by='date' , ascending = True)
df = df.reset_index(drop = True)
df
```

**Intermediate Exercises**

```python
import pandas as pd
import numpy as np

data = {
    'date': ['2024-01-05', '2024-01-18', '2024-02-03', '2024-02-27',
             '2024-03-15', '2024-04-10', '2024-04-22', '2024-05-08',
             '2024-06-14', '2024-07-01', '2024-08-19', '2024-09-30'],
    'category': ['Food', 'Electronics', 'Food', 'Beauty',
                 'Electronics', 'Food', 'Beauty', 'Electronics',
                 'Food', 'Beauty', 'Electronics', 'Food'],
    'revenue': [1500, 8200, 1800, 3200, 9500, 2100, 2800, 7600,
                1900, 3500, 8800, 2200],
    'units_sold': [120, 45, 140, 60, 52, 110, 55, 48, 130, 65, 50, 115]
}
df2 = pd.DataFrame(data)
df2['date'] = pd.to_datetime(df2['date'])
df2
```

1. Extract + group by month

Add a month column, then find the total revenue per month. Reset index cleanly.

```python
df2['month'] = df2['date'].dt.month
df_month_rev = df2.groupby('month')['revenue'].sum().reset_index()
df_month_rev
```

2. Filter by date range

Filter all sales between March 1, 2024 and June 30, 2024 (inclusive). Reset index.

```python
filt = (df2['date'] >= pd.to_datetime('2024-03-01')) & (df2['date'] <= pd.to_datetime('2024-06-30'))
df_march_june = df2[filt].reset_index(drop=True)
df_march_june
```

3. The "client monthly report" scenario

A logistics company wants to know:

*   Total revenue per quarter (total_revenue)
*   Average units sold per quarter (avg_units)



```python
pythondf = df2.set_index('date', inplace=True)
```

```python
# Aggregate quarterly total revenue (summed by Quarter End)
df_monthly_report = df2.resample('QE').agg({
    'revenue': 'sum',
    'units_sold' : 'mean'
})
df_monthly_report
```

A Moroccan agricultural export company wants a full time series analysis of their 2024 shipments.


```python
import pandas as pd
import numpy as np

data = {
    'date': pd.date_range(start='2024-01-01', periods=24, freq='SMS'),
    'product': ['Dates', 'Olive Oil', 'Argan Oil', 'Honey',
                'Dates', 'Olive Oil', 'Argan Oil', 'Honey',
                'Dates', 'Olive Oil', 'Argan Oil', 'Honey',
                'Dates', 'Olive Oil', 'Argan Oil', 'Honey',
                'Dates', 'Olive Oil', 'Argan Oil', 'Honey',
                'Dates', 'Olive Oil', 'Argan Oil', 'Honey'],
    'revenue': [12000, 8500, 15000, 6200, 13500, 9200, 14800, 7100,
                11800, 9800, 16200, 6800, 14200, 8900, 15500, 7300,
                13800, 9500, 15800, 6900, 12500, 8800, 14500, 7200],
    'units_exported': [500, 320, 180, 290, 560, 345, 175, 310,
                       480, 360, 195, 280, 530, 335, 185, 300,
                       510, 350, 190, 295, 490, 340, 182, 305]
}
df3 = pd.DataFrame(data)
```
