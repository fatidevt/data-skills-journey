# Be careful when copying arrays

```python
a = np.array([1,2,3])
b = a
b[0] = 100
print(a)
# the change happens to a and b
```

```python
a = np.array([1,2,3])
b = a.copy()
b[0] = 100
print(a)
# the change happens to b only
```

# Reshaping Arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after = before.reshape((4,2))
print(after)
```

```python
# Vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2])
```

```python
# Horizontal stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.hstack([v1,v2])
```

# Boolean Masking and Advanced Indexing

```python
# Boolean array (mask) showing whether each element satisfies the condition.
filedata > 15
```

```python
# NumPy uses the Boolean mask to select only the values where the answer was True
filedata[filedata > 15]
```

```python
# Returns True for rows with any value > 15
np.any(filedata > 15, axis = 1)
```

```python
# Returns True for columns with any value > 15
np.any(filedata > 15, axis = 0)
```

```python
# 2 conditions
( (filedata > 15) & (filedata < 18) )
```
